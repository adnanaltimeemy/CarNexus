```python
# %% [markdown]
# #  RollsRoyce Sales, Profit Margins & Demographics  UK Dataset
# 
# This notebook gathers sales figures, financial data, and demographic indicators
# related to RollsRoyce in the United Kingdom.

# %%
# --- 1) Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# %%
# --- 2) Create UK Sales Data (from DVLA new registration history) ---
# Based on available UK registration data (DVLA new registrations)
years = list(range(2001, 2026))
sales_uk = [
    102, 90, 97, 165, 162, 183, 240, 236, 180, 371,
    388, 379, 380, 460, 452, 554, 475, 498, 516, 370,
    352, 533, 514, 465, 231
]

rolls_royce_uk_sales_df = pd.DataFrame({
    "Year": years,
    "UK_Sales": sales_uk
})

rolls_royce_uk_sales_df.head()

# %%
# --- 3) Combine Profit Margin Data ---
# Current RollsRoyce Group Gross Profit Margins (all divisions) 
# Note: UK specific margin for cars isn't reported separately publicly.
profit_margin_data = {
    "Metric": ["Gross Profit Margin"],
    "Value (%)": [25.3]  # Approx annual gross margin from recent investing.com data
}

profit_margin_df = pd.DataFrame(profit_margin_data)
profit_margin_df

# %%
# --- 4) Global RollsRoyce Car Sales (Approx including Europe) ---
# Reference global sales for context (not UK only)
global_sales_data = {
    "Year": [2022, 2023, 2024],
    "Global_Sales": [6021, 6032, 5712]  # worldwide deliveries
}

global_sales_df = pd.DataFrame(global_sales_data)
global_sales_df

# %%
# --- 5) Visualize UK Sales Trends ---
plt.figure(figsize=(12,6))
sns.lineplot(data=rolls_royce_uk_sales_df, x="Year", y="UK_Sales", marker='o')
plt.title("RollsRoyce Motor Cars UK Registrations / Sales (20012025)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("New UK Registrations")
plt.show()

# %%
# --- 6) Visualize Global vs UK Sales (if desired) ---
merged = pd.merge(
    rolls_royce_uk_sales_df[rolls_royce_uk_sales_df.Year >= 2022],
    global_sales_df,
    on="Year",
    how="left"
)

plt.figure(figsize=(10,5))
sns.barplot(data=merged.melt(id_vars="Year", var_name="Region", value_name="Units"),
            x="Year", y="Units", hue="Region")
plt.title("UK vs Worldwide RollsRoyce Sales (20222024)", fontsize=14)
plt.ylabel("Units Delivered")
plt.show()

# %%
# --- 7) Demographics Placeholder ---
# Real demographics data may need to be purchased or obtained from luxury market reports;
# here we create a placeholder sample for illustration.

demo_sample = {
    "Demographic_Group": ["Under 40 Buyers", "Age 4060 Buyers", "Over 60 Buyers"],
    "Percentage (%)": [15, 45, 40]  # Example allocations
}

demographics_df = pd.DataFrame(demo_sample)
demographics_df

# %%
# --- 8) Save DataFrames for Export ---
rolls_royce_uk_sales_df.to_csv("rolls_royce_uk_sales.csv", index=False)
profit_margin_df.to_csv("rolls_royce_profit_margin.csv", index=False)
demographics_df.to_csv("rolls_royce_demographics.csv", index=False)
```
