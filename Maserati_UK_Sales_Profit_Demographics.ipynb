```python
# 1) Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2) UK Maserati Annual Sales
# Data source: HowManyCarsMade (DVLA registrations)
maserati_sales_uk = {
    "Year": [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,
             2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,
             2021,2022,2023,2024,2025],
    "Units_Sold": [357,425,349,434,435,394,591,723,421,470,
                   405,316,339,1260,1436,1502,1754,1353,1000,625,
                   795,747,923,490,181]
}

df_sales = pd.DataFrame(maserati_sales_uk)
df_sales.head()
```

```python
import pandas as pd

profit_data = {
    "Year": [2020, 2021, 2022, 2023, 2024],
    "Turnover (£ millions)": [4.7, 5.3, 5.8, 6.2, 6.8],
    "Profit_PreTax (£ thousands)": [176, 200, 250, 321, 326]
}

df_profit = pd.DataFrame(profit_data)

# Calculate profit margin
df_profit["Profit_Margin (%)"] = round(
    (df_profit["Profit_PreTax (£ thousands)"] /
     (df_profit["Turnover (£ millions)"] * 1000)) * 100,
    2
)

df_profit
```

```python
"Profit_PreTax (£ thousands)"
```

```python
# 4) Merge Sales & Profit Data (approximate by year)
df_merged = pd.merge(df_sales, df_profit, how="left", on="Year")
df_merged.tail(10)
```

```python
# 5) Sales Analysis Visualisation
plt.figure(figsize=(12,6))
plt.plot(df_sales["Year"], df_sales["Units_Sold"], marker='o')
plt.title("Maserati UK Annual Sales (Units)")
plt.xlabel("Year")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()
```

```python
# 6) Profit Margin Visualisation
plt.figure(figsize=(8,5))
plt.bar(df_profit["Year"].astype(str), df_profit["Profit_Margin (%)"])
plt.title("Profit Margin % for Maserati UK (2020–2024)")
plt.ylabel("Profit Margin (%)")
plt.show()
```
