```python
# Import key libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Aston Martin UK sales data by year
sales_data = {
    "Year": list(range(2001, 2026)),
    "UK_Sales": [
        548, 551, 495, 686, 1597, 2418, 2223, 1624, 1296,
        1142,1081,1081,1051,994,1129,1184,1667,1705,1944,
        1029,1131,1187,1067,1162,582  # 2025 single provisional estimate
    ]
}

df_sales = pd.DataFrame(sales_data)
df_sales.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_sales["Year"], df_sales["UK_Sales"], marker='o', linestyle='-')
plt.title("Aston Martin UK Sales (20012025)")
plt.xlabel("Year")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()
```

```python
# Profit margin table (example data)
profit_data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Gross_Margin_%": [39.74, 34.45, 18.16, 31.38, 32.62, 39.15, 36.86]
}

df_profit = pd.DataFrame(profit_data)
df_profit
```

```python
plt.bar(df_profit["Year"], df_profit["Gross_Margin_%"], color="navy")
plt.title("Aston Martin Gross Profit Margin (Selected Years)")
plt.xlabel("Year")
plt.ylabel("Gross Margin (%)")
plt.show()
```

```python
# Placeholder demographic data
demo_data = {
    "Customer_ID": range(1, 101),
    "Age_Group": np.random.choice(["18-30","31-45","46-60","60+"], 100),
    "Gender": np.random.choice(["Male","Female","Other"], 100),
    "Region_UK": np.random.choice(
        ["London","South East","Midlands","North England","Scotland","Wales","Northern Ireland"],
        100
    ),
    "Income_Bracket": np.random.choice(
        ["100k-200k","200k-300k","300k-500k","500k+"],
        100
    ),
    "Model_Purchased": np.random.choice(
        ["Vantage","DB11","DBX","DB12","Valhalla"], 100
    )
}

df_demo = pd.DataFrame(demo_data)
df_demo.head()
```

```python
# Count by age group
age_counts = df_demo["Age_Group"].value_counts()
print("Customer Age Group Distribution")
print(age_counts)

# Visualize distribution
age_counts.plot(kind="bar")
plt.title("Aston Martin UK Buyer Age Groups")
plt.show()
```

```python
# Example join (sales trends + hypothetical average income per year)
mean_income_by_year = df_demo.groupby("Age_Group")["Income_Bracket"].count()

# This is an illustrative step
print(mean_income_by_year)
```
