```python
# Step 1 — Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Optional: For nicer charts
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
# Step 2 — Sales Data for UK (example data)
sales_data = pd.DataFrame({
    "Year": [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Units_Sold_UK": [320, 350, 300, 380, 420, 500, 5421]  # 2023/24 UK ownership estimate from industry study
})

sales_data
```

```python
# Step 3 — Global Profitability Example (proxy)
profit_data = pd.DataFrame({
    "Period": ["2024 Full Year", "H1 2025"],
    "Profit_Margin_%": [27.0, 26.6],       # Operating margin approximate based on published figures
    "Total_Sales": [10687, 5681],          # Pax delivered globally
})

profit_data
```

```python
# Step 4 — Example Demographic Data (UK customer profiles)
demo_data = pd.DataFrame({
    "Demographic": ["Age 25-34", "Age 35-44", "Age 45-54", "Age 55+"],
    "Percentage_%": [15, 30, 35, 20],
    "Typical_Income_£": [250000, 350000, 500000, 600000]
})

demo_data
```

```python
# Step 5 — Combine/merge as needed
combined = sales_data.merge(
    profit_data, left_index=True, right_index=True, how="outer"
)

combined
```

```python
# Step 6 — Chart: UK Sales Trend
plt.figure(figsize=(8,5))
sns.lineplot(x="Year", y="Units_Sold_UK", data=sales_data, marker="o")
plt.title("Lamborghini Units Sold in UK — Trend")
plt.ylabel("Units Sold")
plt.xlabel("Year")

plt.show()
```

```python
# Step 7 — Chart: Demographic Breakdown
plt.figure(figsize=(6,6))
plt.pie(demo_data["Percentage_%"], labels=demo_data["Demographic"], autopct='%1.1f%%')
plt.title("Example UK Lamborghini Buyer Age Demographics")
plt.show()
```
