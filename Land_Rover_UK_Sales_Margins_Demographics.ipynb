```python
# LandRover_UK_Sales_Demographics.ipynb

# ---
# ## 1) Imports and Setup
# ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
# 2.1) UK annual sales (new registrations)
uk_sales = pd.DataFrame({
    'Year': [2023, 2024, 2025],
    'Land_Rover_Registrations': [59000, 60500, 60720]  # estimated from best-selling data
})
uk_sales
```

```python
# 2.2) Land Rover average dealer profit margins (used cars)
uk_margins = pd.DataFrame({
    'Model': ['Range Rover Evoque', 'Discovery Sport', 'Other LR Models'],
    'Average_Margin_GBP': [4450, 3671, 3000]  # average margins
})
uk_margins
```

```python
import numpy as np

# 3.1) Simulated buyer age distribution
np.random.seed(100)

ages = np.random.choice(range(25, 75), size=500)
```

```python
# 3.2) Simulated buyer gender labels
genders = np.random.choice(['Male','Female','Other'], size=500, p=[0.65,0.32,0.03])
```

```python
demo = pd.DataFrame({
    'Buyer_Age': ages,
    'Buyer_Gender': genders
})
```

```python
# 4.1) Bar chart — New UK sales over time
plt.figure(figsize=(8, 5))
sns.barplot(x='Year', y='Land_Rover_Registrations', data=uk_sales, palette='dark')
plt.title('Land Rover UK New Registrations (2023–2025)')
plt.ylabel('Units Registered')
plt.xlabel('Year')
plt.show()
```

```python
# 4.2) Boxplot — Dealer Profit Margins
plt.figure(figsize=(6, 5))
sns.boxplot(y='Average_Margin_GBP', data=uk_margins)
plt.title('Land Rover Average Dealer Profit Margins (GBP)')
plt.ylabel('Profit Margin (£)')
plt.show()
```

```python
# 4.3) Buyer demographic distributions
plt.figure(figsize=(8,4))
sns.histplot(demo['Buyer_Age'], bins=10, kde=True, color='teal')
plt.title('Simulated Buyer Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Buyer_Gender', data=demo, palette='pastel')
plt.title('Simulated Buyer Gender Distribution')
plt.show()
```

```python
# Summary combined dataset
sales_margins_demo = pd.merge(
    uk_sales.assign(key=1),
    uk_margins.assign(key=1),
    on='key'
).drop(columns=['key'])

sales_margins_demo.head()
```
