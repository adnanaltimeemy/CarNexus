```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')   # updated style name

uk_sales_data = {
    'Year': list(range(2001, 2026)),
    'New_Registrations': [554, 511, 536, 487, 740, 682, 753, 744, 516, 559,
                           629, 834, 887, 1014, 1486, 1276, 1155, 1167, 1256, 1121,
                           1171, 1085, 1049, 1015, 471]
}

uk_sales_df = pd.DataFrame(uk_sales_data)

plt.figure(figsize=(10,5))
plt.plot(uk_sales_df['Year'], uk_sales_df['New_Registrations'], marker='o')
plt.title('Ferrari New Registrations in UK (2001â€“2025)')
plt.xlabel('Year')
plt.ylabel('New Registrations')
plt.grid(True)
plt.show()
```
